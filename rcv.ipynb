```python
import pandas as pd
import json
```

```python
allVotes = pd.read_csv("allvotes.csv")
```

```python
ranks = ["DEM Mayor Choice 1 of 5 Citywide (026916)",
         "DEM Mayor Choice 2 of 5 Citywide (226916)",
         "DEM Mayor Choice 3 of 5 Citywide (326916)",
         "DEM Mayor Choice 4 of 5 Citywide (426916)",
         "DEM Mayor Choice 5 of 5 Citywide (526916)"]
```

```python
allVotes[ranks[0]].unique().tolist()
```

```python
allVotes["Final Round"] = ""
allVotes["Ranks"] = None
```

```python
allVotes
```

```python
voteRecord = {}
for row in allVotes.itertuples():
    winner = "None"
    oldPrecinct = row[2]
    precinct = oldPrecinct[4:6] + oldPrecinct[-3:]
    ranks = []
    for i in range(3, 8):
        ranks.append(row[i])
    if ("overvote" in ranks):
        ranks = ranks[:ranks.index("overvote")]
    if (ranks.count("undervote") >= 2):
        if (ranks.index("undervote") < 4):
            if (ranks[ranks.index("undervote") + 1] == "undervote"):
                ranks = ranks[:ranks.index("undervote")]
    allVotes.at[row.Index, "Ranks"] = ranks
    voteRecord[precinct + " - " + str(row[1])] = ranks
```

```python
with open('voterecord.json', 'w') as f:
    json.dump(voteRecord, f)
```
